```python
# Import necessary libraries
import os
from llama_index.core import Document, VectorStoreIndex, load_index_from_storage
from llama_index.embeddings.huggingface import HuggingFaceEmbedding
from llama_index.core import Settings, StorageContext
from llama_index.core.node_parser import SentenceSplitter
import PyPDF2
import glob
import torch
from transformers import pipeline

# Define the path to your PDF files
pdf_folder_path = '/content/drive/MyDrive/RAG/Data/*.pdf'

# Create a list to hold Document objects
document_objects = []

# Read each PDF file and create Document objects
for pdf_file in glob.glob(pdf_folder_path):
    with open(pdf_file, 'rb') as file:
        reader = PyPDF2.PdfReader(file)
        text = ''
        for page in reader.pages:
            text += page.extract_text() + '\n'
        document = Document(text=text)
        document_objects.append(document)

# Function to build or load the vector store index
def get_build_index(documents, embed_model="BAAI/bge-small-en-v1.5", save_dir="./vector_store/index"):
    """
    Builds or loads a vector store index from the given documents.

    Args:
        documents (list[Document]): A list of Document objects.
        embed_model (str, optional): The embedding model to use. Defaults to "local:BAAI/bge-small-en-v1.5".
        save_dir (str, optional): The directory to save or load the index from. Defaults to "./vector_store/index".

    Returns:
        VectorStoreIndex: The built or loaded index.
    """
    # Set index settings
    Settings.llm = None  # Disable any LLM settings since you're not using OpenAI
    Settings.embed_model = HuggingFaceEmbedding(model_name="BAAI/bge-small-en-v1.5")
    Settings.node_parser = SentenceSplitter(chunk_size=1000, chunk_overlap=200)
    Settings.num_output = 512
    Settings.context_window = 3900

    # Check if the save directory exists
    if not os.path.exists(save_dir):
        # Create and load the index
        index = VectorStoreIndex.from_documents(documents, service_context=Settings)
        index.storage_context.persist(persist_dir=save_dir)
    else:
        # Load the existing index
        index = load_index_from_storage(
            StorageContext.from_defaults(persist_dir=save_dir),
            service_context=Settings,
        )
    return index

# Get the Vector Index
vector_index = get_build_index(documents=document_objects, embed_model="local:BAAI/bge-small-en-v1.5", save_dir="./vector_store/index")

# Create a query engine from the index
query_engine = vector_index.as_query_engine()

# Define your query
query = "Describe the documents"
response = query_engine.query(query)

# Use a lightweight model for text generation
model_id = "facebook/bart-large-cnn"  # You can replace this with a more suitable lightweight model
pipe = pipeline(
    "text-generation",
    model=model_id,
    torch_dtype=torch.bfloat16,
    device_map="auto",
    max_length=1024  # Increase this value based on your needs
)

# Process the response and generate text
if response.response:
    generated_response = pipe(response.response)[0]['generated_text']
    print("Generated Response:", generated_response)
else:
    print("No response found for the query.")
```
